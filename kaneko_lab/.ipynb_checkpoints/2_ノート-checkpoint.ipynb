```python
import numpy as np
from numpy.random import randn
from matplotlib import pyplot as plt
%matplotlib inline

n = 10
data_count = 300
test_data_count = 100

X = randn(data_count, n)
w = randn(n)
w_star = randn(n)
o_star = np.dot(X, w_star)

test_X = randn(test_data_count, n)
test_o_star = np.dot(test_X, w_star)

def activation_function(X, w):
    return np.dot(X, w) + randn(X.shape[0]) / 100

def loss_function(X, w, o_star):
    return np.sum((activation_function(X, w) - o_star) ** 2)

def nabla_function(X, w, o_star):
    return 2 * np.dot(activation_function(X, w) - o_star, X)

# learn_rateを1e-3より大きくすると、収束せずinfなどに発散する。
def learn_weight(X, w, o_star, learning_count=10000, learning_rate=1e-5):
    learning_step = 0
    loss = np.empty(learning_count)
    while learning_count > learning_step:
        print(loss_function(X, w, o_star)) if learning_step % 100 == 0 else None
        loss[learning_step] = loss_function(test_X, w, test_o_star)
        w -= nabla_function(X, w, o_star) * learning_rate
        learning_step += 1
    plt.plot(loss)
    
learn_weight(X, w, o_star)
```

```python
# L2正則化をかける。正則化項の比率ごとにグラフを見てみる
import copy

X = randn(300, 10)
w = randn(10)
w_star = randn(10)
o_star = np.dot(X, w_star)

test_X = randn(100, 10)
test_o_star = np.dot(test_X, w_star)

_lambda = 1.0

def activation_function(X, w):
    return np.dot(X, w) + randn(X.shape[0]) / 100

def loss_function(X, w, o_star):
    return np.sum((activation_function(X, w) - o_star) ** 2) + _lambda * np.sum(w ** 2)

def nabla_function(X, w, o_star):
    return 2 * np.dot(activation_function(X, w) - o_star, X) + _lambda * 2 * w

# learn_rateを1e-3より大きくすると、収束せずinfなどに発散する。
def learn_weight(X, w, o_star, learning_count=10000, learning_rate=1e-7):
    learning_step = 0
    loss = np.empty(learning_count)
    while learning_count > learning_step:
        print(loss_function(X, w, o_star)) if learning_step % 1000 == 0 else None
        loss[learning_step] = loss_function(test_X, w, test_o_star)
        w -= nabla_function(X, w, o_star) * learning_rate
        learning_step += 1
    plt.plot(loss, label="lambda="+str(_lambda))
    plt.legend()

for _lambda in [0.1, 0.7, 5.0, 32.0, 100.0]:
    _w = copy.deepcopy(w)
    learn_weight(X, _w, o_star)
```

```python
# L2正則化をかける。正則化項の比率ごとにグラフを見てみる
import copy

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X = randn(300, 10)
w = randn(10)
w_star = randn(10)
o_star = sigmoid(np.dot(X, w_star)) + randn(X.shape[0]) / 100

test_X = randn(100, 10)
test_o_star = sigmoid(np.dot(test_X, w_star))

_lambda = 1.0

def activation_function(X, w):
    return sigmoid(np.dot(X, w))

def loss_function(X, w, o_star):
    return np.sum((activation_function(X, w) - o_star) ** 2) + _lambda * np.sum(w ** 2)

def nabla_function(X, w, o_star):
    return 2 * np.dot(activation_function(X, w) - o_star, sigmoid(X) * (1 - sigmoid(X)) * X) + _lambda * 2 * w

# learn_rateを1e-3より大きくすると、収束せずinfなどに発散する。
def learn_weight(X, w, o_star, learning_count=10000, learning_rate=1e-7):
    learning_step = 0
    loss = np.empty(learning_count)
    while learning_count > learning_step:
        print(loss_function(X, w, o_star)) if learning_step % 1000 == 0 else None
        loss[learning_step] = loss_function(test_X, w, test_o_star)
        w -= nabla_function(X, w, o_star) * learning_rate
        learning_step += 1
    plt.plot(loss, label="lambda="+str(_lambda))
    plt.legend()

# 正則化項の比率が大きいほど、lossの初期値は当然大きく、収束した先も比較的大きくなる。
for _lambda in [0.1, 0.7, 5.0, 32.0, 100.0]:
    _w = copy.deepcopy(w)
    learn_weight(X, _w, o_star)
```

```python
# ネットワーク構築
# 後から気づいたが、計算速度的に重みも行列にしてレイヤーをオブジェクトの単位にしたほうが良かった。
import numpy as np
from numpy.random import rand
from numpy.random import randn
import copy

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X = rand(300, 2)
def xor(X):
    _X = copy.deepcopy(X.T)
    _X[0] = X.T[0] > 0.5
    _X[1] = X.T[1] > 0.5
    return np.logical_xor(_X[0], _X[1]).astype(float).T

test_X = rand(100, 2)

class Perceptron:
    def __init__(self, in_size, _lambda=1.0):
        self._in_size = in_size
        self._weight = randn(in_size)
        self._lambda = _lambda
        self._last_X = np.zeros(in_size)
        
    def out(self, X):
        self._last_X = X
        return sigmoid(np.dot(X, self._weight))

    # o_starではなく、スカラにしてデバッグ
    def loss_function(self, o_star, X=None):
        if X == None:
            X = self._last_X
        return np.sum((self.out(X, self._weight) - o_star) ** 2) + \
            self._lambda * np.sum(self._weight ** 2)

    def nabla_w(self, o_star, X=None):
        if X == None:
            X = self._last_X
        return 2 * np.dot(self.out(X) - o_star, 
            sigmoid(X) * (1 - sigmoid(X)) * X) + self._lambda * 2 * self._weight
    
    # ここがちょっと詰まってる
    def nabla_i(self, o_star, X=None):
        if X == None:
            X = self._last_X
        return 2 * np.dot((self.out(X) - o_star).reshape(300, 1), 
            (sigmoid(self._weight) * (1 - sigmoid(self._weight)) * self._weight).reshape(1, self._in_size))
        
    def learn(self, o_star, learning_rate, X=None):
        if X == None:
            X = self._last_X
        self._weight -= self.nabla_w(o_star) * learning_rate
        
    def pre_o_star(self, o_star, X=None):
        if X == None:
            X = self._last_X
        return X - self.nabla_i(o_star)

class Network:
    def __init__(self, in_size, layer_count=2, perceptron_count=10):
        self._layer_count = layer_count
        self._perceptron_count = perceptron_count
        # in_sizeの入力ベクトルをバッチで行列としてを受け取り、入力数分のごとに出力
        # ベクトルをアウトプットするパーセプトロンをperceptron_count分用意し、それ
        # らの出力全てを入力として受け取り、その数分出力をoutputする隠れ層をperceptron_count
        # 全結合する。それらの出力をout_perceptronで受け取って一つの出力を返す
        # ネットワーク。
        self._layeres = [[Perceptron(in_size) for i in range(perceptron_count)]]
        for i in range(layer_count - 1):
            self._layeres.append([Perceptron(perceptron_count) for j in range(perceptron_count)])
        self._out_perceptron = Perceptron(perceptron_count)
        
    def out(self, X):
        _in = X
        for layer in self._layeres:
            _in = np.array([perc.out(_in) for perc in layer]).T
        return self._out_perceptron.out(_in)
    
    def loss_function(self, X, o_star):
        return np.sum((self.out(X) - o_star) ** 2)
    
    def learn(self, X, o_star, test_X, test_o_star, learning_count=10000, learning_rate=1e-7):
        learning_step = 0
        loss = np.empty(learning_count)
        while learning_count > learning_step:
            loss[learning_step] = self.loss_function(test_X, test_o_star)
            training_loss = self.loss_function(X, o_star)
            print(training_loss) if learning_step % 1000 == 0 else None
            self._out_perceptron.learn(o_star, learning_rate)
            _o_star = self._out_perceptron.pre_o_star(o_star)
            for layer in self._layeres:
                _next_o_star = np.zeros(self._perceptron_count)
                for i, perc in enumerate(layer):
                    perc.learn(_o_star[i], learning_rate)
                    _next_o_star += perc.pre_o_star(_o_star[i])
                _o_star = _next_o_star / self._perceptron_count
            learning_step += 1
        
Perceptron(2)

xor_network = Network(2)
xor_network.learn(X, xor(X), test_X, xor(test_X))
```
